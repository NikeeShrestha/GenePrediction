```python
import pandas as pd
import sys
```

```python
def read_fasta(file_name):
    sequences={}
    
    F = open(file_name)
    
    for line in F:
        if line.startswith('>'):
            defline=line.strip()
            defline=defline.split(' ')[0]
            defline=defline.split('>')[1]
            
            if defline not in sequences:
                sequences[defline]=''
        else:
            sequences[defline] += line.strip()
    return(sequences)
```

```python
sequence=read_fasta("RawGene/Zmays_833_Zm-B73-REFERENCE-NAM-5.0.55.cds_primaryTranscriptOnly.fa")
```

```python
new_sequence = {key: {'seq': value, 'CDSLength': len(value)} for key, value in sequence.items()}
```

```python
def GCcount(gene):
    GC=0
    A=0
    G=0
    T=0
    C=0
    
    for element in range(0, len(gene)):
        if str(gene[element])=="G": 
            GC+=1
            G+=1
        elif str(gene[element])=="C":
            GC+=1
            C+=1
        elif str(gene[element])=="A":
            A+=1
        elif str(gene[element])=="T":
            T+=1
    
    GCcontent=(GC/len(gene))*100
    Gcontent=(G/len(gene))*100
    Acontent=(A/len(gene))*100
    Tcontent=(T/len(gene))*100
    Ccontent=(C/len(gene))*100
    
    
    return GCcontent,Gcontent,Acontent,Tcontent,Ccontent
```

```python
for key, value in new_sequence.items():
    value['GCcontent'], value['Gcontent'],value['Acontent'],value['Tcontent'],value['Ccontent']=GCcount(value["seq"])
```
